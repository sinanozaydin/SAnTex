## EBSD Preprocessing

### This line imports the necessary functionality from the Sage library to work with EBSD data.

````python
from sage import EBSD
````

### Here, the code creates an instance of the EBSD class, which represents the EBSD data stored in the file "ebsd.ctf". For this example, ebsd.ctf is located in the directory this notebook is being run from. But you may provide a relative path like data/ebsd.ctf or an absolute import such as /Users/utpal/Documents/ebsd/ebsd.ctf

````python
ebsdfile = EBSD("ebsd.ctf")
````

### This code snippet is used to determine the phases present in the EBSD file along with their respective percentage abundances.

This line calls the phases() method on the ebsdfile object, which likely belongs to the EBSD class. This method returns a dictionary where the keys are the names of different phases present in the EBSD file, and the values are the percentage abundances of those phases.

The output from running this code is a dictionary containing phase names as keys and their corresponding percentage abundances as values. For example:



````python
ebsdfile.phases()
````

## To check the methods present in the object, enter the object name and hit Tab

### This code snippet retrieves data and header information from the EBSD file object for further analysis or visualization.

This line calls the get_ebsd_data() method on the ebsdfile object, likely belonging to the EBSD class. This method returns a DataFrame containing detailed EBSD data, such as phase, coordinates (X and Y), bands, error, Euler angles (Euler1, Euler2, Euler3), mean angular deviation (MAD), and additional columns (BC, BS). Each row represents a point in the sample.

The output DataFrame looks like this:


````python
ebsdfile.get_ebsd_data()
````

### This DataFrame provides detailed information about each point in the sample, including its phase, position, and various measurement parameters.

This line calls the get_ebsd_data_header() method, which returns a tuple containing two elements:

A DataFrame with the header information of the EBSD data, including the column names and their meanings.
Additional information such as the number of cells in the X and Y directions, the step size in the X and Y directions, and acquisition Euler angles.
The output tuple looks like this:


````python
ebsdfile.get_ebsd_data_header()
````

#### The first DataFrame contains column names and their meanings, while the second DataFrame provides additional information about the data acquisition setup.



# To get the euler angles of a particular phase, the syntax is

## Function: `get_euler_angles`

### Description:
This function retrieves the Euler angles of a specific phase from the EBSD file.

### Parameters:
- `phase_number` (int): The index of the phase for which Euler angles are to be retrieved. 

### Returns:
- `euler_angles` (numpy.ndarray): An array containing the Euler angles (in radians) of the specified phase.

### Usage Example:
```python
nan = ebsdfile.get_euler_angles(0)
print(nan)


````python
nan = ebsdfile.get_euler_angles(0)
nan
````

````python
Forsterite = ebsdfile.get_euler_angles(1)
````

### Returns:
- `ds` (DataFrame): A downsampled version of the EBSD data stored in a DataFrame.

### Usage Example:
```python
